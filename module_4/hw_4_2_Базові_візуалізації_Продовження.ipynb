### Завдання: Візуалізація функцій експоненти, логарифму, і модифікованої синусоїди

**Мета:** Створити фігуру з трьома панелями: на першій показати графік експоненціальної функції, на другій - логарифмічну функцію, а на третій - модифіковану синусоїду.

1. Імпортуйте необхідні бібліотеки:


```python
from matplotlib import pyplot as plt
import numpy as np
```


2. Використовуючи фукнціонал `numpy` створіть масив `x` з 400 точок в діапазоні між -2 та 2 і масив `x_positive` в діапазоні між 0.1 і 2 також з 400 точок. Другий масив в нас буде для логарифмічної функції, перший з більш широким діапазоном - для інших функцій.


```python
x = np.linspace(-2, 2, 400)
y = np.linspace(0.1, 2, 400)
```


3. Тут маємо кілька кроків.

  3.1. Ініціалізуйте фігуру Matplotlib з розміром, достатнім для трьох панелей (рекомендую спробувати експериментально виявити добрий розмір).   
  3.2. На першій панелі візуалізуйте експоненціальну функцію `y = e^x`. Використовуйте відповідну змінну `x`.  
  3.2. На другій панелі покажіть логарифмічну функцію `y = ln(x)`.  
  3.4. На третій панелі покажіть модифіковану синусоїду `y = sin(x) + x`.  

  Відобразіть графік.


```python
# Спосіб 1 - використовуючи звичайний plt

plt.figure(figsize = (20, 5))

plt.subplot(1, 3, 1)
plt.plot(x, np.exp(x))

plt.subplot(1, 3, 2)
plt.plot(y, np.log(y))

plt.subplot(1, 3, 3)
plt.plot(x, np.sin(x) + x)

plt.show();
```

```python
# Спосіб 2 - використовуючи Object Oriented Interface

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(20, 5)

ax[0].plot(x, np.exp(x))
ax[1].plot(y, np.log(y))
ax[2].plot(x, np.sin(x) + x);
```

4. Модифікація і довершення графіку.  
  4.1. Для кожної панелі встановіть назву графіка та підписи до осей х та у (можна так і назвати `x` та формула функції).  
  4.2. Використайте різні кольори для кожного графіка.  


```python
# Спосіб 1 - використовуючи звичайний plt

plt.figure(figsize = (20, 5))

plt.subplot(1, 3, 1)
plt.plot(x, np.exp(x), color='red')
plt.title('Функція експоненти')
plt.xlabel('x')
plt.ylabel('e^x')

plt.subplot(1, 3, 2)
plt.plot(y, np.log(y), color='blue')
plt.title('Функція логарифму')
plt.xlabel('y')
plt.ylabel('ln(x)')

plt.subplot(1, 3, 3)
plt.plot(x, np.sin(x) + x, color='green')
plt.title('Модифікована синусоїда')
plt.xlabel('x')
plt.ylabel('sin(x) + x')

plt.show();
```

```python
# Спосіб 2 - використовуючи Object Oriented Interface

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(20, 5)

ax[0].plot(x, np.exp(x), color='red')
ax[0].set(xlabel='x', ylabel='e^x',
       title='Функція експоненти')

ax[1].plot(y, np.log(y), color='blue')
ax[1].set(xlabel='y', ylabel='ln(x)',
       title='Функція логарифму')

ax[2].plot(x, np.sin(x) + x, color='green')
ax[2].set(xlabel='x', ylabel='sin(x) + x',
       title='Модифікована синусоїда');
```

5. А тепер зобразіть ті самі 3 трафіки тих самих кольорів тільки на одному графіку. Визначіть назву кожної з ліній для легенди та обовʼязково відобразіть легенду.

```python
# Спосіб 1 - використовуючи звичайний plt

fig = plt.figure()
plt.plot(x, np.exp(x), color='red', label='e^x')
plt.plot(y, np.log(y), color='blue', label='ln(x)')
plt.plot(x, np.sin(x) + x, color='green', label='sin(x) + x')
plt.title('Функції експоненти,  Функція логарифму і Модифікована синусоїда')
plt.xlabel('x / y')
plt.ylabel('f(x) / f(y)')
plt.legend();
```

```python
# Спосіб 2 - використовуючи Object Oriented Interface

ax = plt.axes()

ax.plot(x, np.exp(x), color='red', label='e^x')
ax.plot(y, np.log(y), color='blue', label='ln(x)')
ax.plot(x, np.sin(x) + x, color='green', label='sin(x) + x')
ax.set_title('Функції експоненти,  Функція логарифму і Модифікована синусоїда')
ax.set_xlabel('x / y')
ax.set_ylabel('f(x) / f(y)')
ax.legend();
```
