6. Трансформуйте колонки 'Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків' у тип даних float. УВАГА! Перед трансформацією подивіться, які дані містять ці колонки. Аби успішно трансформувати тип даних, треба спочатку привести рядки до того виду, який буде підходити, аби потім рядок зробити типом float.

```python
# Трансформація атрибуту 'Ваш вік' - Спосіб 1
df['Ваш вік'] = df['Ваш вік'].apply(lambda x: x.replace(',', '.')).astype(float)
df['Ваш вік'].value_counts()
```

```python
# Трансформація атрибуту 'Зарплата у $$$ за місяць, лише ставка після сплати податків' - Спосіб 2
df['Зарплата у $$$ за місяць, лише ставка після сплати податків'] = df['Зарплата у $$$ за місяць, лише ставка після сплати податків'].fillna('0').apply(lambda x: x.replace(',', '.')).astype(float)
```

```python
df.dtypes
```

```python
#Спосіб 2

cols = ['Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків']

for col in cols:
    if col == 'Ваш вік':
        df[col] = df[col].str.replace(',', '.').astype(float)
    else:
        df[col] = df[col].fillna('0').str.replace(',', '.').astype(float)
```

```python
df.dtypes
```

    Для цього треба імпортувати бібліотеку `datetime`. Я пропоную написати функцію для обробки кожного окремого рядка, яка враховує довжину рядка і на основі цього застосовує різний формат даних дати.

```python
# 1. Виведення на екран перших і останніх 5-ти вартостей атрибуту Timestamp

display(
    df['Timestamp'].head(),
    df['Timestamp'].tail()
)
```

```python
# 2. Перевірка форматів даних
df['Timestamp'].str.len().unique()
```

```python
# Cпосіб 1 - якщо по кожній окремій вартості в конкретній колонці
import datetime

def transform_timestamps(value):
    if len(value) > 16:
        return datetime.datetime.strptime(value, '%m.%d.%Y %H:%M:%S')
    else:
        return datetime.datetime.strptime(value, '%m.%d.%Y %H:%M')
```

```python
df['Timestamp'] = df['Timestamp'].apply(transform_timestamps)
```

```python
df.dtypes
```

```python
# Cпосіб 2 - якщо по всіх рядках відразу в заданій колонці
import datetime

df['timestamp_len'] = df.Timestamp.str.len()

def process_timestamp(row):
    if row.timestamp_len > 16:
        return datetime.datetime.strptime(row.Timestamp, '%m.%d.%Y %H:%M:%S')
    else:
        return datetime.datetime.strptime(row.Timestamp, '%d.%m.%Y %H:%M')
```

```python
df['Timestamp'] = df.apply(process_timestamp, axis=1)
```

```python
df.dtypes
```
